```python
import json
import re

def count_words_in_json(json_data):
    # 将JSON对象转换为字符串
    json_string = json.dumps(json_data)

    # 使用正则表达式提取单词
    words = re.findall(r'\b\w+\b', json_string)

    # 统计单词数量
    word_count = len(words)

    return word_count

def write_dataframe_to_json(df, word_limit, output_file_prefix='output_file'):
    # 将表格数据转换为字典格式
    data = df.to_dict(orient='records')

    # 初始化变量
    current_word_count = 0
    current_chunk_data = []
    chunk_number = 1

    # 分段写入JSON文件
    for record in data:
        # 将字典转换为 JSON 字符串
        json_data = json.dumps([record],ensure_ascii=False, indent=4)

        # 统计单词数量
        record_word_count = count_words_in_json(record)

        # 如果当前块加上当前记录的单词数量超过限制，则写入新文件
        if current_word_count + record_word_count > word_limit:
            # 构建输出文件名
            output_file_name = f'{output_file_prefix}_{chunk_number}.json'

            # 将当前块写入输出文件
            with open(output_file_name, 'w') as f:
                f.write(json.dumps(current_chunk_data, indent=4))

            # 重置变量
            current_word_count = 0
            current_chunk_data = []
            chunk_number += 1

        # 添加记录到当前块
        current_word_count += record_word_count
        current_chunk_data.append(record)

    # 处理最后一个块
    if current_chunk_data:
        output_file_name = f'{output_file_prefix}_{chunk_number}.json'
        with open(output_file_name, 'w') as f:
            f.write(json.dumps(current_chunk_data, indent=4))

# 示例使用
# 假设 df_cleaned 是您的 DataFrame，每个文件包含的单词数量限制为20
write_dataframe_to_json(df_cleaned, 20000)
```

```python
from string import Template
```

```python
def getSummary(key):
  template_summary = Template('''
{
"参数": "$data",
"需求规格": "",
"供应商反馈": ""
}
  ''')
  return template_summary.substitute(data=key)
```

```python
def getContent(data,key):
  template_content = Template('''
这是数据：
$data
----------
请回复如下格式：
[{
"参数":"",
"需求规格":"",
""供应商反馈""：""
}]
你的任务：
提取$key的参数规格需求信息和供应商反馈信息。
下面是你的回复(使用英文)：
''')
  return template_content.substitute(data=data,key=key)
```

```python

for index, row in pd_read.iterrows():
    para = row['parameter']
    data = row['json_data']
    if(not pd.isnull(data)):
      pd_read.at[index, 'content'] = getContent(data,para)
      pd_read.at[index, 'summary'] = getSummary(para)
```
